```python
# basic indexing 
# basic indexing in numpy will be exactly like python list but all in one [] seperated with ","
# start:end:step ==> start and end in arrays just like iterables in pure python are [)
np_arr0 = np.arange(25).reshape(5,5)
np_arr0
```

```python
# first index ==> axis 0 (rows) , second index ==> axis 1 (colums), third index ==> axis 2 (depth, channels , bands) , ...
np_arr0[:,-1] # all of the last column
```

```python
np_arr0[0,:] # all of the first row
```

```python
np_arr0[2:4, 1:4] # rows index from 2 to 3, columns indexed from 1 to 3
```

```python
np_arr0[:,1::2] # the columns placed in odd indices 
```

```python
np_arr0[..., -1] # equalivent of np_arr0[:,-1] / more useful on matrices with higher dimesnion and is called ellipsis
```

```python
# advanced indexing, indexing with conditional statements 
np_arr1 = np.arange(-10, 10).reshape(5,4)
np_arr1
```

```python
# np will check the condition, elements by elements and replace the values by results 
np_arr1 > 0
```

```python
# meaning return only the values that are replaced by True
np_arr1[np_arr1 > 0]
```

```python
# making a random matrix of int that we only want the odd numbers
np.random.seed(0) #ignore for now 
np_rand0 = np.random.randint(10, 100, (3,3))
np_rand0
```

```python
np_rand0[np_rand0 % 2 ==1]
```

```python
# fancy indexing
# in fancy indexing we give the array exactly the indices we want and it will return them
```

```python
np_rand1 = np.random.randint(10, 100, (8, 8))
np_rand1
```

```python
np_rand1[np.array([1, 2, 4])]  # return the second, third and fourth rows
```

```python
# in fact this is a combination of fancy and basic indexing
np_rand1[:,np.array([1, 2, 4])]  # return the second, third and fourth columns
```

```python
np_rand1[np.array([1, 5, 2]), np.array([3, 6, 4])] # returning the elements with these indices (1,3) ,(5,6), (2,4)
```

<h2 style="color:#6153CC"> Dimension manipulation</h2>


```python
# with help of new axis cmd in np we can add exactly one dimension to our array
one_d_arr = np.array([1, 2, 3, 4])
one_d_arr.shape
```

```python
two_d_arr = one_d_arr[:, np.newaxis ]
two_d_arr.shape
```

```python
# how to expand dimensions even more?
# combining with use of tuples
twelve_d_arr = np.expand_dims(one_d_arr, axis=tuple(range(1,12)))
twelve_d_arr.shape
```

<h2 style="color:#6153CC"> Concatenation </h2>


```python
# vstack, hstack, concanate
# vstack , vertically stacking two arrays
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
np.vstack([arr1, arr2])
```

```python
#hstack , horizontally stacking 
np.hstack([arr1, arr2])
```

```python
#concanate, concatenate in any axis 
np.concatenate([arr1, arr2], axis=0)
```

```python
# how to concat in 2nd axis 
# we first need to give our data a second and third dimension hence the first and second axis 
arr1 = arr1[:, np.newaxis, np.newaxis]
arr2 = arr2[:, np.newaxis, np.newaxis]
```

```python
# then use concat command to concat them 
np.concatenate([arr1, arr2], axis=2)
```

```python
imshow(elonBGR)
```

```python
imshow(elonRGB)
```
